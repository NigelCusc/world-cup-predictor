# 2026 Group Standings Extraction

Pulls **group standings tables** (position, team, played, W/D/L, goals, points) from the [2026 FIFA World Cup](https://en.wikipedia.org/wiki/2026_FIFA_World_Cup) Wikipedia page.

Output: `Data/group_standings_2026.pkl` — a dictionary keyed by `Group A` … `Group L`.

This is separate from match-level scraping in `scripts/scrape_fixtures_2026.py`.

```python
import pickle
from pathlib import Path

import pandas as pd

WIKIPEDIA_URL = "https://en.wikipedia.org/wiki/2026_FIFA_World_Cup"
REQUEST_HEADERS = {"User-Agent": "FifaWorldCupAnalysis/1.0 (educational data project)"}
OUTPUT_PATH = Path("Data/group_standings_2026.pkl")
GROUP_LETTERS = list("ABCDEFGHIJKL")  # 12 groups in 2026 (48-team format)
```

```python
all_tables = pd.read_html(WIKIPEDIA_URL, storage_options=REQUEST_HEADERS)
print(f"Loaded {len(all_tables)} tables from Wikipedia")
```

```python
def is_group_standings_table(df: pd.DataFrame) -> bool:
    """Detect Wikipedia group standings tables (4 teams, Pld/Pts columns)."""
    if len(df) != 4 or len(df.columns) < 8:
        return False
    cols = " ".join(str(c).lower() for c in df.columns)
    return "pld" in cols and "pts" in cols and "team" in cols


standings_indices = [i for i, t in enumerate(all_tables) if is_group_standings_table(t)]
print(f"Found {len(standings_indices)} group standings tables at indices: {standings_indices}")

if len(standings_indices) != len(GROUP_LETTERS):
    raise ValueError(
        f"Expected {len(GROUP_LETTERS)} group tables, found {len(standings_indices)}. "
        "Wikipedia page layout may have changed."
    )
```

```python
def normalize_team_name(name: str) -> str:
    """Strip Wikipedia host marker, e.g. 'Mexico (H)' -> 'Mexico'."""
    return name.replace(" (H)", "").strip()


dict_table = {}
for letter, table_index in zip(GROUP_LETTERS, standings_indices):
    df = all_tables[table_index].copy()
    team_col = next(c for c in df.columns if "team" in str(c).lower())
    df = df.rename(columns={team_col: "Team"})
    df["Team"] = df["Team"].apply(normalize_team_name)
    dict_table[f"Group {letter}"] = df

list(dict_table.keys())
```

```python
for group_name, df in dict_table.items():
    print(f"===== {group_name} =====")
    display(df)
```

```python
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
with open(OUTPUT_PATH, "wb") as output:
    pickle.dump(dict_table, output)

print(f"Saved {len(dict_table)} group tables to {OUTPUT_PATH}")
```
